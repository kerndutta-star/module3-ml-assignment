```python
# Exercise 3: Self-Reflection Prompt for Improving Output
# Runs with a deterministic mock LLM so it always executes.

import textwrap

# ----------------------------
# Mock LLM (no API key needed)
# ----------------------------
def mock_llm(system_prompt: str, user_prompt: str) -> str:
    u = user_prompt.lower()

    # Initial summary
    if "summarize the text below" in u:
        return (
            "Acme Mobile released a battery-life update, but some users report more app crashes. "
            "The company is investigating, suggests restarting and updating apps, and plans a hotfix within a week. "
            "Users can contact support if issues continue."
        )

    # Self-reflection + improved version
    if "critique it using these criteria" in u:
        return textwrap.dedent("""
        1) Self-Critique:
        - The summary is accurate but slightly wordy.
        - It uses three sentences instead of the requested maximum of two.
        - Some phrasing could be more concise for readability.

        2) Revised Summary:
        Some users say Acme’s battery-life update is causing more app crashes.
        The company is investigating, recommends restarting and updating apps,
        plans a hotfix within a week, and advises contacting support if problems continue.
        """).strip()

    return "OK"

def call_llm(system_prompt: str, user_prompt: str) -> str:
    return mock_llm(system_prompt, user_prompt)

# ----------------------------
# Source Text
# ----------------------------
source_text = """
Acme Mobile launched an update to improve battery life, but some users report increased app crashes.
Acme says it is investigating and recommends updating apps and restarting the phone.
A hotfix is planned within a week. Users can contact support if crashes continue.
"""

# ----------------------------
# STEP 1 — Generate Initial Summary
# ----------------------------
SYSTEM_SUM = "You are a precise summarizer."
PROMPT_SUM = f"""
Summarize the text below in 3 sentences maximum, plain language.

TEXT:
{source_text}
"""

before_summary = call_llm(SYSTEM_SUM, PROMPT_SUM)

# ----------------------------
# STEP 2 — Self-Reflection Prompt
# ----------------------------
SYSTEM_REFLECT = "You are a careful editor. First critique, then revise."
PROMPT_REFLECT = f"""
Here is a summary:

{before_summary}

Critique it using these criteria:
- Accuracy: no new facts
- Clarity: simple wording
- Audience: non-technical
- Length: max 2 sentences
- Must include: (1) problem, (2) company response, (3) next step

Then provide:
1) Self-critique (bullets)
2) Revised summary (max 2 sentences)
"""

after_output = call_llm(SYSTEM_REFLECT, PROMPT_REFLECT)

# ----------------------------
# CLEAN OUTPUT PRINTING
# ----------------------------
print("\n==============================")
print("EXERCISE 3: SELF-REFLECTION OUTPUT")
print("==============================")

print("\nSTEP 1 — Original Summary (BEFORE)")
print(before_summary)

print("\nSTEP 2 — Self-Critique + Improved Summary (AFTER)")
print(after_output)
```
