````python
import re
import traceback

# 1) Build the ReACT prompt (what you show in your submission)
REACT_PROMPT = f"""
You must follow an iterative ReACT cycle:

[REASON]
Briefly restate the task and give a short plan.

[ACT:CODE]
Return ONLY a Python code block that:
- Uses the pandas DataFrame df (already defined)
- Converts df['date'] to datetime
- Groups revenue by month
- Prints a table with columns month, total_revenue
- Plots a line chart (matplotlib)
- Includes basic try/except error handling

[ACT:RUN]
Do NOT run anything yourself.

Data preview:
{df.head().to_string(index=False)}
""".strip()

print("=== REACT PROMPT ===")
print(REACT_PROMPT)

# 2) Call the mock model
raw_response = mock_react_llm(REACT_PROMPT)

print("\n=== MODEL RESPONSE (REASON + CODE) ===")
print(raw_response)

# 3) Extract Python code from ```python ... ```
match = re.search(r"```python\s*(.*?)```", raw_response, flags=re.DOTALL | re.IGNORECASE)
if not match:
    raise ValueError("No Python code block found in the model response.")

generated_code = match.group(1).strip()

print("\n=== EXTRACTED CODE (what will run) ===")
print(generated_code)

# 4) Execute generated code and show output
print("\n=== EXECUTION OUTPUT ===")
try:
    exec(generated_code, globals(), locals())
    print("\nExecution succeeded ✅")
except Exception:
    print("\nExecution failed ❌")
    print(traceback.format_exc())
````
